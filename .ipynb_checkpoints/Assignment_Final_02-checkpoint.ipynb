# 1. Introduction

随着上世纪全球气候变化的影响，人们逐渐开始重视化石能源的过量使用以及寻找其可替代品这一问题。迄今为止，煤炭、石油和天然气等化石燃料是造成全球气候变化的最大因素，占全球温室气体排放量的 75% 以上，占所有二氧化碳排放量的近 90%。（引用）因此，进入21世纪以来，世界上很多国家通过积极协商和制定减少碳排放的政策，持续加大对可再生能源行业的研究与应用投入力度，来积极应对气候变化带来的全球挑战。

在此背景下，本项研究着眼于21世纪以后，可再生能源利用领域这一全球能源政策的核心部分中，哪些因素对全球各国人均国内生产总值有重要的影响。同时，基于部分文献，尝试从经济因素、可再生能源领域因素、人口和土地面积对全球各国进行类型划分。

此研究可以为制定促进可再生能源产业的扶持与发展政策提供数据支持和理论依据，为未来的全球可再生能源发展方向做出积极探索。

**237英文**

With the effects of global climate change over the last century, attention has gradually begun to focus on the excessive use of fossil energy sources and the search for alternatives to them. Fossil fuels, such as coal, oil and gas, are by far the largest contributor to global climate change, accounting for over 75 percent of global greenhouse gas emissions and nearly 90 percent of all carbon dioxide emissions. **(Ritchie, Roser & Rosado, 2020)** Therefore, since the beginning of the 21st century, many countries have been actively responding to the global challenges posed by climate change by actively negotiating and formulating policies to reduce carbon emissions and continuing to increase their investment in the research and application of the renewable energy sector.

Against the backdrop of these global challenges, this study takes a comprehensive approach, examining the various factors in the renewable energy utilization sector, a key component of global energy policy, that significantly influence GDP per capita in countries worldwide beyond the 21st century. Furthermore, drawing from existing literature, the study attempts to categorize  global countries based on economic factors, renewable energy sector factors, population, and land area, thereby enhancing the credibility and thoroughness of the research.

This research can provide data support and a theoretical basis for the formulation of policies to promote the support and development of the renewable energy industry and make positive explorations for the future direction of global renewable energy development.








# 2. Literature review

Tudor 和 Sova（2021 年）将可再生能源消费作为因变量，试图确定一个国家的经济发展、能源效率和创新活动是否会影响其消费向绿色能源的转变。有鉴于此，大量文献探讨了其他因素如何影响可再生能源的发展模式。相比之下，本研究认为，新能源的发展与经济的发展应该是双向因果关系，可以相互作用，而不是一方决定另一方。因此，本研究以世界各国的人均 GDP 为因变量，以可再生能源部门因素和国家因素为自变量，建立研究模型。

**131英文**

Tudor and Sova (2021) used renewable energy consumption as the dependent variable in an attempt to determine whether the economic development of a country, the energy efficiency, and its innovative activities) impact its consumption shift towards green energy sources. 

With this in mind, a large body of literature has explored how other factors have influenced patterns of renewable energy development. In contrast, the current study argues that the development of new energy sources and the development of the economy should have a two-way causal relationship that can interact with each other rather than one determining the other. Therefore, this study uses the world countries' GDP per capita as the dependent variable and the renewable energy sector factors and national factors as the independent variables to build a model for the study.

### 5.1.2 Using VIF to deal with multicollinearity

```python
# calculating VIF
# This function is adjusted from: https://stackoverflow.com/a/51329496/4667568
from statsmodels.stats.outliers_influence import variance_inflation_factor 
from statsmodels.tools.tools import add_constant

def drop_column_using_vif_(df, thresh=5):
    df = df.replace({True: 1, False: 0}) # convert bool to int
    '''
    Calculates VIF each feature in a pandas dataframe, and repeatedly drop the columns with the highest VIF
    A constant must be added to variance_inflation_factor or the results will be incorrect

    :param df: the pandas dataframe containing only the predictor features, not the response variable
    :param thresh: (default 5) the threshould VIF value. If the VIF of a variable is greater than thresh, it should be removed from the dataframe
    :return: dataframe with multicollinear features removed
    '''
    while True:
        # adding a constatnt item to the data. add_constant is a function from statsmodels (see the import above)
        df_with_const = add_constant(df)

        vif_df = pd.Series([variance_inflation_factor(df_with_const.values, i) 
               for i in range(df_with_const.shape[1])], name= "VIF",
              index=df_with_const.columns).to_frame()

        # drop the const
        vif_df = vif_df.drop('const')
        
        # if the largest VIF is above the thresh, remove a variable with the largest VIF
        # If there are multiple variabels with VIF>thresh, only one of them is removed. This is because we want to keep as many variables as possible
        if vif_df.VIF.max() > thresh:
            # If there are multiple variables with the maximum VIF, choose the first one
            index_to_drop = vif_df.index[vif_df.VIF == vif_df.VIF.max()].tolist()[0]
            print('Dropping: {}'.format(index_to_drop))
            df = df.drop(columns = index_to_drop)
        else:
            # No VIF is above threshold. Exit the loop
            break

    return df
```

### 5.2.4 XGBoost model

In this part, we will use the XGBoost method to predict "GDP_per_capita". We will follow the same workflow as RF.

XGBoost has two groups of hyperparameters:

1. The hyperparameters of decision trees, such as max_depth, min_samples_split;
2. The configurations of XGBoost, such as the number of trees (called **n_estimators**) and regularisation hyperparameters.

```python
# values of max_depth and min_samples_split
hyperparameters = {'max_depth':[10,20,30,40,50], 'n_estimators':[50,100,150,200,250]}

randomState_xgb = 125
xgb = XGBRegressor(random_state=randomState_xgb)

# cv=5 by default, which means 5-fold cross-validation
gscv_xgb = GridSearchCV(xgb, hyperparameters)

gscv_xgb.fit(train_x, train_y)

# we can query the best parameter value and its accuracy score
print ("The best parameter value is: ")
print (gscv_xgb.best_params_)
print ("The best score is: ")
print (gscv_xgb.best_score_)
```

上述结果表明，Random Forest模型与XGBoost模型表现相差不大，都好于Regression Tree模型.

对于模型生产的结果原因有以下几方面：

1. 本次研究的因变量是"GDP_per_capita"，在数据集中，自变量不仅包含了可再生能源领域的因素，同时包含能源领域的通用指标，如能效强度，人均通电量等。这些指标与人均GDP更为相关。

2. 在能源种类方面，我们可以发现，化石能源在人均GDP影响因素的重要性仍然很高。这与事实相吻合：尽管新能源产业在21世纪高速发展，但化石能源仍然在世界能源体系中占主导地位。

3. 在可再生能源方面，人均年水力发电量非常重要。这与报告中的结论相一致：目前，水力发电的发电量超过所有其他可再生能源技术的总和，预计到 2030 年代仍将是世界上最大的可再生能源发电来源。https://www.iea.org/energy-system/renewables/hydroelectricity

### 4.5.1 在总数据集中，影响最大的还是"Primary_E_consumption_kWh_per_capita" 人均一次能源消费（千瓦时/人）：每个人平均消费的一次性能源量 和 "Energy_intensity_level_of_primary_energy" 一次能源的能源强度水平（兆焦/2017年购买力平价GDP美元）：能源消耗量与产出（GDP）之间的比率，反映了经济活动的能源效率；在此之后则是通电率。

```python
def mapping_clusters(labels_cluster):
    gdf['cluster_nm'] = labels_cluster
    gdf.plot(column='cluster_nm', categorical=True, legend=True, figsize=(12,8), cmap='Paired');
```

```python
# adapted from this tutorial: https://towardsdatascience.com/how-to-make-stunning-radar-charts-with-python-implemented-in-matplotlib-and-plotly-91e21801d8ca
def radar_plot_cluster_centroids(df_cluster_centroid):
    # parameters
    # df_cluster_centroid: a dataframe with rows representing a cluster centroid and columns representing variables
    
    # add an additional element to both categories and restaurants that’s identical to the first item
    # manually 'close' the line
    categories = df_cluster_centroid.columns.values.tolist()
    categories = [*categories, categories[0]]
    
    label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(categories))
    
    plt.figure(figsize=(12, 8))
    plt.subplot(polar=True)
    for index, row in df_cluster_centroid.iterrows():
        centroid = row.tolist()
        centroid = [*centroid, centroid[0]]
        label = "Cluster {}".format(index)
        plt.plot(label_loc, centroid, label=label)
    plt.title('Cluster centroid comparison', size=20, y=1.05)
    lines, labels = plt.thetagrids(np.degrees(label_loc), labels=categories)
    plt.legend()
    plt.show()
```

# 分级聚类

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, leaf_rotation=90., **kwargs)
```

```python
agg_cluster = AgglomerativeClustering(distance_threshold=0, n_clusters=None).fit(normed)
ax = plt.gca()
plt.title("Hierarchical Clustering Dendrogram")
# plot the top three levels of the dendrogram
plot_dendrogram(agg_cluster, truncate_mode="level", p=3)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.ylabel('distance')
# note that these two red dashed lines are drawn manually, as we see that the gap between the two lines is larger than other gaps
plt.hlines(4.0, ax.get_xlim()[0], ax.get_xlim()[1], linestyle='dashed', color='r')
plt.hlines(2.8, ax.get_xlim()[0], ax.get_xlim()[1], linestyle='dashed', color='r')
plt.hlines(2.4, ax.get_xlim()[0], ax.get_xlim()[1], linestyle='dashed', color='r')
plt.show()
```
