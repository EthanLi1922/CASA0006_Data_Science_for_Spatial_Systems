# 1. Introduction

Fossil fuels, such as coal, oil and gas, are by far the largest contributor to global climate change, accounting for over 75 percent of global greenhouse gas emissions. (Ritchie, Roser & Rosado, 2020) Therefore, many countries have been actively responding to this global challenges by formulating policies to reduce carbon emissions and continuing to increase investment in the research and application of the renewable energy.

In this context, this study takes a comprehensive approach, examining the various factors in the renewable energy utilization sector which significantly influence GDP per capita in countries worldwide. Furthermore, the study attempts to categorize global countries based on economic indicators, renewable energy indicators and other indicators, thereby enhancing the credibility and thoroughness of the research. This research can provide data support and a theoretical basis for the formulation of policies to promote the renewable energy industry and make positive explorations for the future of renewable energy development.

# 2. Literature review

Tudor and Sova (2021) used renewable energy consumption as the dependent variable in an attempt to determine whether the economic development of a country, the energy efficiency, and its innovative activities) impact its consumption shift towards green energy sources. Meanwhile, Shahbaz, M. et al. (2020) conducted a study on renewable energy consumption and economic growth in 38 countries based on previous studies and categorised the results as positive impact, insignificant, and negative impact based on the results. This suggests that promoting new energy sources can have different economic benefits depending on the category of countries. As an essential reference for Shahbaz, M, the essay written by Bhattacharya et al. (2016) pre-listed representative countries according to energy categories and categorized them to study the impact of renewable energy consumption on economic growth. Despite the different data indicators, these articles still provide important reference ideas. 

Based on these research studies, our study argues that there is a complex causal relationship between new energy and economic development. This study will use supervised learning to build a regression model with GDP per capita as the dependent variable and unsupervised learning to conduct a cluster analysis of countries.

# Resaerch question 

1. What factors in the renewable energy use field significantly impact GDP per capita on a global scale?

2. How can similar global energy consumption patterns be identified based on characteristics such as a country's economic indicators, energy consumption structure, population, and land area?

### 5.1.2 Using VIF to deal with multicollinearity

```python
# calculating VIF
# This function is adjusted from: https://stackoverflow.com/a/51329496/4667568
from statsmodels.stats.outliers_influence import variance_inflation_factor 
from statsmodels.tools.tools import add_constant

def drop_column_using_vif_(df, thresh=5):
    df = df.replace({True: 1, False: 0}) # convert bool to int
    '''
    Calculates VIF each feature in a pandas dataframe, and repeatedly drop the columns with the highest VIF
    A constant must be added to variance_inflation_factor or the results will be incorrect

    :param df: the pandas dataframe containing only the predictor features, not the response variable
    :param thresh: (default 5) the threshould VIF value. If the VIF of a variable is greater than thresh, it should be removed from the dataframe
    :return: dataframe with multicollinear features removed
    '''
    while True:
        # adding a constatnt item to the data. add_constant is a function from statsmodels (see the import above)
        df_with_const = add_constant(df)

        vif_df = pd.Series([variance_inflation_factor(df_with_const.values, i) 
               for i in range(df_with_const.shape[1])], name= "VIF",
              index=df_with_const.columns).to_frame()

        # drop the const
        vif_df = vif_df.drop('const')
        
        # if the largest VIF is above the thresh, remove a variable with the largest VIF
        # If there are multiple variabels with VIF>thresh, only one of them is removed. This is because we want to keep as many variables as possible
        if vif_df.VIF.max() > thresh:
            # If there are multiple variables with the maximum VIF, choose the first one
            index_to_drop = vif_df.index[vif_df.VIF == vif_df.VIF.max()].tolist()[0]
            print('Dropping: {}'.format(index_to_drop))
            df = df.drop(columns = index_to_drop)
        else:
            # No VIF is above threshold. Exit the loop
            break

    return df
```

```python
# adapted from this tutorial: https://towardsdatascience.com/how-to-make-stunning-radar-charts-with-python-implemented-in-matplotlib-and-plotly-91e21801d8ca
def radar_plot_cluster_centroids(df_cluster_centroid):
    # parameters
    # df_cluster_centroid: a dataframe with rows representing a cluster centroid and columns representing variables
    
    # add an additional element to both categories and restaurants thatâ€™s identical to the first item
    # manually 'close' the line
    categories = df_cluster_centroid.columns.values.tolist()
    categories = [*categories, categories[0]]
    
    label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(categories))
    
    plt.figure(figsize=(12, 8))
    plt.subplot(polar=True)
    for index, row in df_cluster_centroid.iterrows():
        centroid = row.tolist()
        centroid = [*centroid, centroid[0]]
        label = "Cluster {}".format(index)
        plt.plot(label_loc, centroid, label=label)
    plt.title('Cluster centroid comparison', size=20, y=1.05)
    lines, labels = plt.thetagrids(np.degrees(label_loc), labels=categories)
    plt.legend()
    plt.show()
```

**Use the Elbow method to find the 'right' number of clusters**

The idea behind the Elbow method is to identify the value of k where the distortion begins to decrease most rapidly. At this value, if you continue to increase the k, you will not get much decrease on the SSE.

```python
# calculate SSE for a range of number of cluster
list_SSE = []
min_k = 1
max_k = 10
range_k = range(min_k, max_k+1)
for i in range_k:
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(normed)
    # inertia is a concept from physics. Roughly it means SSE of clustering.
    list_SSE.append(km.inertia_)

# plot
plt.plot(range_k, list_SSE, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

According to this figure, although the rate of decline is fast for k=2, this gives the same results as DBSCAN and is not very interpretable. At the same time, since the rate of decline does not slow down significantly until k=4, k=3 can be considered as an â€˜elbow point.

### 5.4.5 Hierarchical clustering

Next, we will use Agglomerative Clustering (one type of hierarchical clustering) for our research. This method helps us to explore clustering at different levels without specifying the number of clusters in advance.

We can generate and draw a dendrogram of the clustering process, which shows the hierarchical structure of the clusters and helps us to decide the appropriate number of clusters.

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, leaf_rotation=90., **kwargs)
```

```python
agg_cluster = AgglomerativeClustering(distance_threshold=0, n_clusters=None).fit(normed)
ax = plt.gca()
plt.title("Hierarchical Clustering Dendrogram", size=15, color='black', y=1.1)
# plot the top three levels of the dendrogram
plot_dendrogram(agg_cluster, truncate_mode="level", p=3)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.ylabel('distance')
# note that these two red dashed lines are drawn manually, as we see that the gap between the two lines is larger than other gaps
plt.hlines(4.00, ax.get_xlim()[0], ax.get_xlim()[1], linestyle='dashed', color='r')
plt.hlines(2.67, ax.get_xlim()[0], ax.get_xlim()[1], linestyle='dashed', color='r')
plt.hlines(2.20, ax.get_xlim()[0], ax.get_xlim()[1], linestyle='dashed', color='r')
plt.hlines(2.10, ax.get_xlim()[0], ax.get_xlim()[1], linestyle='dashed', color='r')
plt.hlines(1.23, ax.get_xlim()[0], ax.get_xlim()[1], linestyle='dashed', color='r')
plt.show()
```

According to the results of the dendrogram of the hierarchical clustering, the number of clusters is 2, 3, and 5 all have large gaps. The red dotted line highlights the distance gap. This distance gap in the dendrogram indicates that the next merge may not happen.

However, this correlation does not imply causality. There is a positive correlation between GDP per capita and renewable energy consumption, but it is limited to countries in the upper middle, high and very high income levels. (Tudor, C. and Sova, R, 2021) The increase in solar and wind power generation may be due to greater energy demand due to economic growth. Alternatively, economic growth may have financed the development and application of renewable energy technologies. 

## 6.2 Results - Clustering

In the part of the clustering study, three different methods, DBSCAN, K-means, and hierarchical clustering were used to discuss the cases where the number of clusters was 2, 3, and 5. 

The results of hierarchical clustering are used here as a reference:

Cluster 4 is the first to be separated, and it can be basically determined that this cluster belongs to developing countries. These countries usually have a low GDP per capita, have little capital investment in the new energy field.

Cluster 2 is characterized by having the largest territorial area: Canada, the United States, China, Australia, and Brazil (Russia has no data). These countries have higher per capita solar power generation, which may be related to application scenarios where large solar photovoltaic panels are installed.

Cluster 3 only has Norway and Iceland, characterized by a tiny territory but the highest GDP per capita, as well as a very high renewable energy consumption capacity. This is not only related to economic capacity, but also to the willingness to protect the environment and the special conditions of the country. For example, geothermal resources play a significant role in Iceland's energy supply. (Piao, S., 2023)

Cluster 1 has four countries: Bahrain, Kuwait, Qatar, and Singapore. This clustering is characterized by the smallest territory, a very high GDP per capita, and the highest annual fossil energy generation per capita. The first three countries are Gulf countries with abundant oil reserves. Singapore is too small to import energy, and natural gas accounts for 95 percent of total energy.

Cluster 0 has 25 countries, predominantly in Western Europe, but also including Japan and New Zealand. It is characterized by a high GDP per capita and high per capita solar and wind power generation. Most of these countries are developed countries with high technical and economic capacity to explore the renewable energy field.

## 6.3 Discussion

While the present study has reached some conclusions, the actual conclusions of the study are somewhat sketchy and require the necessary breakthroughs and innovations.

The current study can be used as a starting point for extended research. For example, in the clustering study, Eastern and Western Europe in Europe belong to different clusters, and the energy patterns between them can be compared. It could also focus on whether countries such as China and the United States, which have large land areas, large total populations, strong economies, and high energy consumption, could learn from the experiences of other countries in renewable energy development.

# 7. Conclusion

Our analysis shows that the impact of fossil energy on GDP per capita is still strongly correlated. Hydroelectricity is strongly correlated with GDP per capita in renewable energy. This is corroborated by the International Energy Agency (IEA) report. On the other hand, solar and wind power generation has grown rapidly over the past decade, is strongly correlated with per capita GDP growth, and may have complex causal links to technological advances and affordability. However, the development of new energy technologies is still confined to upper-middle-income and higher-income countries, and most of the world's developing countries do not feature prominently in the new energy sector.

The level of green energy in the energy mix of individual countries remains highly heterogeneous. As a result, renewable energy sources should be developed according to the different conditions of each country. At the same time, developed countries need to invest funds in developing countries to  contribute to energy conservation and emission reduction.
